```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
file_path = 'Entropy_Training-ML.csv'
df = pd.read_csv(file_path)
print(df.head())  
```

```python
#Clean the data
def update_value(entry):
    if isinstance(entry, str) and 'Entropy for' in entry:
        value = float(entry.split(': ')[1])
        return value
    else:
        return entry

# Apply the function to all columns
df = df.applymap(update_value)
print(df)
```

```python
df
```

```python
# Convert columns to type float
df['H'] = df['H'].astype(float)
df['Y'] = df['Y'].astype(float)
```

```python
X = df['H'].values
y = df['Y'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
# Reshape X_train and X_test to have a shape of (n_samples, n_features)
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)

y_train = np.array(y_train)
y_test = np.array(y_test)

print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

Random Forest Classifier: 

```python
# Initialize and train a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)  
rf_classifier.fit(X_train, y_train)
```

```python
X_test
```

```python
# Make predictions on the testing set
y_pred = rf_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
y_pred
```

K-Nearest Neighbors (KNN) Classifier:


```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize and train a KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)  
knn_classifier.fit(X_train, y_train)

# Make predictions on the testing set
y_pred2 = knn_classifier.predict(X_test)
print('X Test:', X_test)
print('Y Pred:', y_pred2)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred2)

print("KNN Classifier Accuracy:", accuracy)
```

Support Vector Machine (SVM) Classifier:

```python
from sklearn.svm import SVC
# Initialize and train a SVM classifier
svm_classifier = SVC(kernel='rbf')  
svm_classifier.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = svm_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)

print("SVM Classifier Accuracy:", accuracy)
```
